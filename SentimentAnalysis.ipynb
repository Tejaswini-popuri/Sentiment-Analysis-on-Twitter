```python
file = open('Trump.txt','r') 
tweets = file.read()
```

## Data Cleaning

As we start cleaning data, we use regex library to clean special characters,urls,usernames, string"b'RT  'and punctuaction marks. Then proceed with converting all words in the list to lowercase as python is case sensitive. finally we store the cleaned data into a list as different elements.

```python
import regex as re
```

```python
#remove links from the data
tweets = re.sub(r"http\S+", " ", tweets)
```

```python
#remove usernames from the data
tweets = re.sub(r"@\S+", " ", tweets)
```

```python
#convert to lower case
tweets = tweets.lower()
```

```python
#remove 'RT  '
tweets = tweets.replace("b'rt  ",' ')
tweets = tweets.replace('nbrt ',' ')
tweets = tweets.replace('brt  ',' ')
```

```python
def normalize_text(tweets):
    
    tweets = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", tweets)
    tweets = re.sub(r"\'s", " \'s", tweets)
    tweets = re.sub(r"\'ve", " \'ve", tweets)
    tweets = re.sub(r"n\'t", " n\'t", tweets)
    tweets = re.sub(r"\'re", " \'re", tweets)
    tweets = re.sub(r"\'d", " \'d", tweets)
    tweets = re.sub(r"\'ll", " \'ll", tweets)
    tweets = re.sub(r",", " , ", tweets)
    tweets = re.sub(r"!", " ! ", tweets)
    tweets = re.sub(r"\(", " \( ", tweets)
    tweets = re.sub(r"\)", " \) ", tweets)
    tweets = re.sub(r"\?", " \? ", tweets)
    tweets = re.sub(r"\s{2,}", " ", tweets)
    return tweets.strip().lower()
```

```python
normalize_text(tweets)
```

```python
#remove special characters
tweets = re.sub('[?!@#$%^&*()_+-\;–./\'’|‘]',' ', tweets)
tweets = re.sub(r'(@[A-Za-z0–9_]+)|[^\w\s\n]|#|http\S+',' ', tweets)
```

```python
tweets = tweets.split()
```

```python
for i in tweets:
    if i=="xe" or i=='x'or i =='xa':
        tweets.remove(i)
    elif i=='n' or i== 'b'or i=='xf'or i=='f':
        tweets.remove(i)
    elif i=='xb'or i=='xac'or i=='xbb'or i=='rt':
        tweets.remove(i)
```

```python
print (tweets)
```

## Open and read the files with Positive, Negative and stop words

 Storing the words from the files in separate lists as pwords, nwords and swords after opening and reading them.

```python
p = open("positive.txt")
pwords= p.read().split()

n = open("negative.txt") 
nwords= n.read().split()
    
s = open("stopwords.txt") 
swords= s.read().split()
```

### Counting the number of Positive ,Negative and Stopwords from Tweets list  

```python
word_count=len(tweets)
print("The word count for the extracted document is "+ str(word_count))
```

```python
pcount=0
ncount=0
scount=0
for i in tweets:
    if i in pwords :
        pcount=pcount+1
    elif i in nwords:
        ncount=ncount+1
    elif i in swords:
        scount=scount+1
print("The word count for the positive words in the document is "+ str(pcount))
print("The word count for the negative words in the document is "+ str(ncount))
print("The word count for the stop words in the document is "+ str(scount))
```

```python
count_ow= word_count-(pcount+ncount+scount) #other word count
pword_ratio = float(pcount/word_count)
nword_ratio = float(ncount/word_count)
sword_ratio = float(scount/word_count)
other_word_ratio = float(count_ow/word_count)
p_to_n_ratio = float((pcount)/(ncount))

print("The ratio of positive words to the whole word count is " +str(pword_ratio))
print("The ratio of negative words to the whole word count is " +str(nword_ratio))
print("The ratio of stop words to the whole word count is " +str(sword_ratio))
print("The ratio of other words to the whole word count is " +str(other_word_ratio))
print("=============================================================")
print("positive to negative ratio is " + str(p_to_n_ratio))
```

## Sentiment Analysis

Finding the difference between positive and negative word count determines the value of sentiment for the keyword.

```python
dif_pcount_ncount = (pcount-ncount)

print ("The Value of sentiment for the word 'Trump' is " +str(dif_pcount_ncount))
```

Percentage difference of positive and negative words gives the overall sentiment for the keyword

```python
p_percent = ((pcount/word_count)*100)
n_percent = ((ncount/word_count)*100)
dif_percent = (p_percent - n_percent)
 
print ("Percentage of positive words is "+ str(p_percent))
print ("Percentage of negative words is "+ str(n_percent))

print ("difference of positive and negative percentages is "+ str(dif_percent))
```

As the difference between positive and negative percentages is negative and not significant the overall sentiment incline towards *weakly negative*

**The above analysis was performed with out the word 'trump' in the positive word list. When 'trump' was included in the positive word list, and the difference of positive and negative percentage was calculated the value was positive due to an increase in the positive word count, thus resulting in a overall sentiment as weakly positive**
